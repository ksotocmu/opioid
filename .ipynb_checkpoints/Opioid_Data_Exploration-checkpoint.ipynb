### Data Set: Emergency Medical Services (EMS) Naloxone Dose Administered (2018 - Current) Quarterly by County
https://data.pa.gov/Opioid-Related/Emergency-Medical-Services-EMS-Naloxone-Dose-Admin/wst4-3int

This dataset contains information on the number of doses given for Naloxone throughout the Commonwealth of Pennsylvania by County. If the Incident County FIPS Code column is blank it means Not Recorded, as this field was blank by the Patient Care Report (PCR) vendor or the EMS Provider. Data is updated Quarterly starting with the first Quarter of Calendar Year 2018.


```python
#Read Naloxone Dataset
nx_data = pd.read_csv('Emergency_Medical_Services__EMS__Naloxone_Dose_Administered_CY_2018_-_Current_Quarterly_County_Health.csv')
pd.read_csv('Emergency_Medical_Services__EMS__Naloxone_Dose_Administered_CY_2018_-_Current_Quarterly_County_Health.csv')
```

```python
nx_data_date = nx_data["Event Date"]
```

```python
#Examine distribution by county
nx_county_dict = {}

for name in nx_data["Incident County Name"]:
    if str(name) in nx_county_dict: nx_county_dict[str(name)] += 1
    else: nx_county_dict[str(name)] = 1
        
print(nx_county_dict)
```

```python
fig=plt.figure(dpi=200)
plt.title("EMS Naloxone Uses by County, 2018-Present") 
plt.xlabel("County") 
plt.ylabel("Uses") 
plt.xticks([])
plt.bar(list(nx_county_dict.keys()),list(nx_county_dict.values()),color='red') 

plt.show()
```

```python
#Sort by date
nx_data['Event Date'] = pd.to_datetime(nx_data['Event Date'])
nx_data = nx_data.sort_values(by='Event Date')
display(nx_data)
```

```python
#Examine distribution by date
nx_date_dict = {}

for date in nx_data["Event Date"]:
    if str(date) in nx_date_dict: nx_date_dict[str(date)] += 1
    else: nx_date_dict[str(date)] = 1

nx_date_list = sorted(nx_date_dict.items())
x,y = zip(*nx_date_list)
```

```python
fig=plt.figure(dpi=200)
plt.title("EMS Naloxone Uses by Date, 2018-Present") 
plt.xlabel("Date: 01/01/2018 - 03/31/2020") 
plt.ylabel("Uses") 
plt.plot(x,y,color='red') 
plt.xticks([])

plt.show()
```

### Data Set: Overdose Information Network Data (2018 - Current) Monthly County State Police
https://data.pa.gov/Opioid-Related/Overdose-Information-Network-Data-CY-January-2018-/hbkk-dwy3

This dataset contains summary information on overdose responses and naloxone administrations by Pennsylvania criminal justice agencies and some third-party (i.e. EMS, Fire, Medical Staff, etc) first responders voluntarily entering incident data. Due to the voluntary nature of the application, the ODIN information provided may not represent the totality of all overdose and/or naloxone administration incidents involving criminal justice agencies occurring within the Commonwealth. Although this dataset does include some third-party administrations of naloxone, it should not be used to measure overdose response and naloxone administration incidents among all first responders.


```python
#Read Police Dataset
police_data = pd.read_csv('Overdose_Information_Network_Data_CY_January_2018_-_Current_Monthly_County_State_Police.csv')
display(police_data)
```

```python
police_data['Incident Date'] = pd.to_datetime(police_data['Incident Date'])
police_data.sort_values(by='Incident Date')
```

```python
#Examine distribution by date
police_date_dict = {}

for date in police_data["Incident Date"]:
    if str(date) in police_date_dict: police_date_dict[str(date)] += 1
    else: police_date_dict[str(date)] = 1
```

```python
police_date_list = sorted(police_date_dict.items())
x,y = zip(*police_date_list)
print(x,y)
```

```python
fig = plt.figure(dpi=150)
plt.title('Police Naloxone Uses by Date')
plt.xlabel('Date: 01/01/2018 - 06/14/2020')
plt.ylabel('Uses')
plt.xticks([])

plt.plot(list(x),list(y))
plt.show()
```

```python
#Examine distribution by day
police_day_dict = {}

for day in police_data["Day"]:
    if str(day) in police_day_dict: police_day_dict[str(day)] += 1
    else: police_day_dict[str(day)] = 1
        
print(police_day_dict)
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
uses = [2325, 2478, 2431, 2588, 3000, 3186, 2666]

fig=plt.figure(dpi=120)
plt.title("Police Naloxone Uses by Day, 2018-Present") 
plt.xlabel("Day") 
plt.ylabel("Uses") 
plt.bar(days,uses) 

plt.show()
```

```python
#Examine distribution by age
police_age_dict = {}

for age in police_data["Age Range"]:
    if str(age) in police_age_dict: police_age_dict[str(age)] += 1
    else: police_age_dict[str(age)] = 1
        
print(police_age_dict)
```

```python
fig=plt.figure(dpi=120)
plt.title("Police Naloxone Uses by Age Group, 2018-Present") 
plt.xlabel("Age Group") 
plt.ylabel("Uses") 
plt.bar(sorted(police_age_dict.keys()),police_age_dict.values()) 

plt.show()
```

### Data Set: Emergency Department Visits for Overdose (Q3 2016 - Current) Quarterly by County
https://data.pa.gov/Opioid-Related/Emergency-Department-ED-Visits-for-Overdose-Identi/vvvs-z9dr


```python
#Load Emergency Department Data
ed_data = pd.read_csv('Emergency_Department__ED__Visits_for_Overdose_Identified_Through_Syndromic_Surveillance_Quarter_3_2016_-_Current_Quarterly_County_Health.csv')
pd.read_csv('Emergency_Department__ED__Visits_for_Overdose_Identified_Through_Syndromic_Surveillance_Quarter_3_2016_-_Current_Quarterly_County_Health.csv')
```

```python
#Examine Lehigh County Opioid ODs by Quarter

lehigh_ed_dict = {}

for num, row in ed_data.iterrows():
    if row['County Name']=="Lehigh" and row['Overdose Type']=="Any Opioid Overdose":
        lehigh_ed_dict[str(row['Time Period'])] = row['Quarterly Rate']

print(lehigh_ed_dict)
```

```python
lehigh_ed_list = sorted(lehigh_ed_dict.items())
x,y = zip(*lehigh_ed_list)

fig = plt.figure(dpi=150)
plt.title("Lehigh County Emergency Department Opioid Visits by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Visits per 100,000")

plt.plot(x,y,color='red')
plt.show()
```

### Generating Javascript for Site 

```python
#Create JS code for Death Rate Graph
for county in sorted(set(mcd_data["County"])):
    print('{label: \'',county[:-4],'\', fill: false, data: [',end='')
    tempList = list(mcd_dict[county]).copy()
    for idx in range(len(tempList)):
        if not math.isnan(tempList[idx]): 
            print(tempList[idx], end='')
        if idx != len(tempList)-1: print(',',end='')
    print('],},')
    
```

```python
#Emergency Department Visits for Opioid OD by County
ed_data = ed_data.sort_values(by="Time Period")

'''
ed_dict: Dictionary of Emergency Dept. Opioid OD Rate Time Series per County

- Sorted chronologically by county

ex. ed_dict["Lehigh"] = {2016 Q3: 76.4, 2016 Q4: 54.3, ...}
'''

ed_dict = {}

county_names = set(ed_data["County Name"])
county_names.remove("Pennsylvania")
county_names = sorted(county_names)

for name in county_names:
    ed_dict[name] = {}

for num, row in ed_data.iterrows():
    if row['Overdose Type']=="Any Opioid Overdose" and row["County Name"] != "Pennsylvania":
        ed_dict[str(row["County Name"])][row["Time Period"]] = row['Quarterly Rate']
```

```python
#Create JS code for ED Graph
for county in county_names:
    print('{label: \'',county,'\', fill: false, data: [',end='')
    tempList = list(ed_dict[county].values()).copy()
    for idx in range(len(tempList)):
        if not math.isnan(tempList[idx]): 
            print(tempList[idx], end='')
        if idx != len(tempList)-1: print(',',end='')
    print('],},')
```
