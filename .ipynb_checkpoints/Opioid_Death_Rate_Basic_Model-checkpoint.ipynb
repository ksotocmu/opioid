## Combine CDC and Census Data:

I combine county-level estimates for income, poverty and employment metrics taken from the U.S. Census Bureau's American Community Survey (1-year estimates). Data is taken for all available years (2010-2018).

#### Income Data:
- <b>Mean Household Income:</b> Mean income (dollars)!!Estimate!!All households (S1902_C02_001E or S1902_C03_001E)

#### Poverty Data:
- <b>% of Individuals below Poverty Line:</b> Percent below poverty level!!Estimate!!Population for whom poverty status is determined (S1701_C03_001E)

#### Employment Data:
- <b>Unemployment Rate (16+):</b> Unemployment rate!!Estimate!!Population 16 years and over (S2301_C04_001E)
- <b>Labor Force Participation Rate (16+):</b> In labor force!!Estimate!!Population 16 years and over (S2301_C02_001E)

```python
main_path = 'Annual Data'
annual_data = pd.read_csv(f'{main_path}.csv')

display(annual_data)
```

```python
county_data_dict = {}
'''
county_data_dict["Example County"] = {2010: [Mean Income, % of Individuals below Poverty Level, Unemployment Rate, 
Labor Force Participation Rate, ...], 2011: [], 2012: [], ...}
'''

for county in set(annual_data["County"]):
    county_data_dict[county[:-4]] = {}
```

```python
#Add Mean Household Income
data_dir = "./Income/"
for year in range(2010,2019):
    temp_data = pd.read_csv(f'{data_dir}ACSST1Y{year}.S1902_data_with_overlays_2020-07-17T095952.csv')
    
    #Select mean household income: S1902_C03_001E
    for idx,row in temp_data.iterrows():
        if idx != 0:
            county_name = row["NAME"][:-14]
            if year < 2017:
                county_data_dict[county_name][year] = [int(row["S1902_C02_001E"])]
            else:
                county_data_dict[county_name][year] = [int(row["S1902_C03_001E"])]
```

```python
#Add Poverty Data
data_dir = "./Poverty/"
for year in range(2010,2019):
    temp_data = pd.read_csv(f'{data_dir}ACSST1Y{year}.S1701_data_with_overlays_2020-07-17T100313.csv')
    
    #Select % of individuals below poverty line: S1701_C03_001E
    for idx,row in temp_data.iterrows():
        if idx != 0:
            county_name = row["NAME"][:-14]
            county_data_dict[county_name][year].append(float(row["S1701_C03_001E"]))
```

```python
#Add Employment Data
data_dir = "./Employment/"
for year in range(2010,2019):
    temp_data = pd.read_csv(f'{data_dir}ACSST1Y{year}.S2301_data_with_overlays_2020-07-21T113120.csv')
    
    #Select Unemployment rate (16+): S2301_C04_001E
    #Select Labor Force Participation Rate (16+): S2301_C02_001E
    for idx,row in temp_data.iterrows():
        if idx != 0:
            county_name = row["NAME"][:-14]
            county_data_dict[county_name][year].append(float(row["S2301_C04_001E"]))
            county_data_dict[county_name][year].append(float(row["S2301_C02_001E"]))
```

```python
print(county_data_dict)
```

## Add Census Data to Annual Data (csv)

```python
for idx,row in annual_data.copy().iterrows():
    if row["Year"] in county_data_dict[row["County"][:-4]]:
        income,poverty,unemployment,labor = county_data_dict[row["County"][:-4]][row["Year"]]
        annual_data.at[idx,"Mean Household Income"] = income
        annual_data.at[idx,"Percent of Individuals Below Poverty Level"] = poverty
        annual_data.at[idx,"Unemployment Rate"] = unemployment
        annual_data.at[idx,"Labor Force Participation Rate"] = labor
```

```python
display(annual_data)
```

```python
#Save Data
annual_data.to_csv(f'{main_path}.csv')
```

## Build Death Rate Model

This is a basic model exploring the relationship between census data and opioid death rate

```python
basic_data = annual_data.copy()

basic_features = ["Year","Mean Household Income","Percent of Individuals Below Poverty Level","Unemployment Rate","Labor Force Participation Rate"]

#Include unreliable rates
for idx,row in annual_data.iterrows():
    if row["Crude Rate"] == "Unreliable":
        basic_data.at[idx,"Crude Rate"] = (float(basic_data.at[idx,"Crude Rate Lower 95% Confidence Interval"]) + float(basic_data.at[idx,"Crude Rate Upper 95% Confidence Interval"]) / 2)

#Drop NaN values for census features
basic_data = basic_data.dropna(subset=basic_features)

#Drop Suppressed/Unreliable values for Death Rate
basic_data = basic_data[basic_data["Crude Rate"] != "Suppressed"]
basic_data = basic_data[basic_data["Crude Rate"] != "Unreliable"]


display(basic_data)
```

```python
#Compare Unemployment Rate and Death Rate

plt.scatter(basic_data["Unemployment Rate"],basic_data["Crude Rate"],c=basic_data["Year"])
plt.title("Unemployment Rate vs. Death Rate")
plt.xlabel("Unemployment Rate")
plt.ylabel("Death Rate")
plt.show()
```

```python
#Compare Labor Force Participation Rate and Death Rate

plt.scatter(basic_data["Labor Force Participation Rate"],basic_data["Crude Rate"],c=basic_data["Year"])
plt.title("Labor Force Participation Rate vs. Death Rate")
plt.xlabel("Labor Force Participation Rate")
plt.ylabel("Death Rate")
plt.show()
```

```python
#Select prediction target
y = basic_data["Crude Rate"]
```

```python
#Define regression features
basic_features = ["Year","Population","Mean Household Income","Percent of Individuals Below Poverty Level","Unemployment Rate","Labor Force Participation Rate"]
X = basic_data[basic_features]
```

```python
X.describe()
```

```python
#Create Training, Test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 12)
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()

# Fit model
reg.fit(X_train, y_train)

#Evaulate accuracy
y_pred = reg.predict(X_test)
print('R^2: ', sklearn.metrics.r2_score(y_test,y_pred))
print('MSE: ', sklearn.metrics.mean_squared_error(y_test,y_pred))
print('Explained Variance: ', sklearn.metrics.explained_variance_score(y_test,y_pred))
```

## Tensorflow Death Rate Regression Model

```python
import tensorflow as tf
import seaborn as sns
```

```python
tf.random.set_seed(12)
```

```python
#Create training, test sets
from sklearn.model_selection import train_test_split

features = ["Population","Year","Unemployment Rate","Labor Force Participation Rate"]


train_data, test_data, train_labels, test_labels = train_test_split(basic_data[features], basic_data["Crude Rate"], test_size=0.2, random_state=12)
```

```python
#Visualize data
sns_plot = sns.pairplot(train_data[features],diag_kind="kde")
#sns_plot.savefig("basic_sns.png")
```

```python
train_stats = train_data.describe()
train_stats = train_stats.transpose()
train_stats
```

```python
#Feature Normalization
def norm(x):
    return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_data)
normed_test_data = norm(test_data)
```

```python
#Model
def build_model():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(64,activation='relu',input_shape=[len(train_data.keys())]),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(1)
    ])
    
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
    
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
EPOCHS = 1000

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(normed_train_data, train_labels, epochs=EPOCHS, validation_split = 0.15, verbose=0, callbacks=[early_stop])
```

```python
hist = pd.DataFrame(history.history)
display(hist)

plt.plot(hist["loss"])
plt.plot(hist['val_loss'],color='orange')
plt.legend()
plt.ylabel("Loss")
plt.xlabel("Epoch")

plt.show()
```

```python
#Make Predictions
test_predictions = model.predict(normed_test_data).flatten()

a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [Death Rate]')
plt.ylabel('Predictions [Death Rate]')
lims = [0, 50]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
#plt.savefig("DeathRateTestError.png",dpi=150)
```

```python
error = test_predictions - test_labels
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [Death Rate]")
_ = plt.ylabel("Count")
```

```python
#Evaulate accuracy
print('R^2: ', sklearn.metrics.r2_score(test_labels,test_predictions))
print('MSE: ', sklearn.metrics.mean_squared_error(test_labels,test_predictions))
print('Explained Variance: ', sklearn.metrics.explained_variance_score(test_labels,test_predictions))
```
