## Combine Emergency Dept and Census Data:

I combine county-level estimates for income, poverty and employment metrics taken from the U.S. Census Bureau's American Community Survey (1-year estimates). Data is taken for all available years (2010-2018).

#### Income Data:
- <b>Mean Household Income:</b> Mean income (dollars)!!Estimate!!All households (S1902_C02_001E or S1902_C03_001E)

#### Poverty Data:
- <b>% of Individuals below Poverty Line:</b> Percent below poverty level!!Estimate!!Population for whom poverty status is determined (S1701_C03_001E)

#### Employment Data:
- <b>Unemployment Rate (16+):</b> Unemployment rate!!Estimate!!Population 16 years and over (S2301_C04_001E)
- <b>Labor Force Participation Rate (16+):</b> In labor force!!Estimate!!Population 16 years and over (S2301_C02_001E)

```python
main_path = 'Annual Data'
annual_data = pd.read_csv(f'{main_path}.csv')

display(annual_data)
```

### <i>Census data has already been added to Annual Data (csv), we can skip this section unless there is a new census observation to add.</i>

```python
county_data_dict = {}
'''
county_data_dict["Example County"] = {2010: [Mean Income, % of Individuals below Poverty Level, Unemployment Rate, 
Labor Force Participation Rate, ...], 2011: [], 2012: [], ...}
'''

for county in set(annual_data["County"]):
    county_data_dict[county[:-4]] = {}
```

```python
#Add Mean Household Income
data_dir = "./Income/"
for year in range(2010,2019):
    temp_data = pd.read_csv(f'{data_dir}ACSST1Y{year}.S1902_data_with_overlays_2020-07-17T095952.csv')
    
    #Select mean household income: S1902_C02_001E or S1902_C03_001E
    for idx,row in temp_data.iterrows():
        if idx != 0:
            county_name = row["NAME"][:-14]
            if year < 2017:
                county_data_dict[county_name][year] = [int(row["S1902_C02_001E"])]
            else:
                county_data_dict[county_name][year] = [int(row["S1902_C03_001E"])]
```

```python
#Add Poverty Data
data_dir = "./Poverty/"
for year in range(2010,2019):
    temp_data = pd.read_csv(f'{data_dir}ACSST1Y{year}.S1701_data_with_overlays_2020-07-17T100313.csv')
    
    #Select % of individuals below poverty line: S1701_C03_001E
    for idx,row in temp_data.iterrows():
        if idx != 0:
            county_name = row["NAME"][:-14]
            county_data_dict[county_name][year].append(float(row["S1701_C03_001E"]))
```

```python
#Add Employment Data
data_dir = "./Employment/"
for year in range(2010,2019):
    temp_data = pd.read_csv(f'{data_dir}ACSST1Y{year}.S2301_data_with_overlays_2020-07-21T113120.csv')
    
    #Select Unemployment rate (16+): S2301_C04_001E
    #Select Labor Force Participation Rate (16+): S2301_C02_001E
    for idx,row in temp_data.iterrows():
        if idx != 0:
            county_name = row["NAME"][:-14]
            county_data_dict[county_name][year].append(float(row["S2301_C04_001E"]))
            county_data_dict[county_name][year].append(float(row["S2301_C02_001E"]))
```

## Add ED Data to Annual Data

```python
ed_dict = {}

for county in set(ed_data["County Name"]):
    ed_dict[county] = {2016:[],2017:[],2018:[],2019:[],2020:[]}
```

```python
ed_data = ed_data.dropna(subset=["Quarterly Rate"])

for idx,row in ed_data.iterrows():
    if row["Type of Rate"] == 'Rate of ED Visits per 10,000 Visits (by facility location)':
        rate = row["Quarterly Rate"]
        ed_dict[row["County Name"]][row["Year"]].append(rate)
```

```python
#print(ed_dict)
```

```python
#Add average yearly ED Visit rates to Annual Data
for idx,row in annual_data.copy().iterrows():
    if row["Year"] in range(2016,2020):
        county_name = row["County"][:-11]
        if len(ed_dict[county_name][row["Year"]]) != 0:
            annual_data.at[idx,"ED Visit Rate"] = statistics.mean(ed_dict[county_name][row["Year"]])
```

## Build Overdose Rate Model

This is a basic model exploring the relationship between census data and opioid overdose ED visits

```python
basic_data = annual_data.copy()

basic_features = ["Population","Year","Mean Household Income","Percent of Individuals Below Poverty Level","Unemployment Rate","Labor Force Participation Rate"]

#Drop NaN values for census features
basic_data = basic_data.dropna(subset=basic_features)

#Drop NaN values for ED Visit Rate
basic_data = basic_data.dropna(subset=["ED Visit Rate"])


display(basic_data)
```

```python
#Compare Labor Force Participation Rate and ED Visit Rate

plt.scatter(basic_data["Labor Force Participation Rate"],basic_data["ED Visit Rate"],c=basic_data["Year"])
plt.title("Labor Force Participation Rate vs. ED Visit Rate")
plt.xlabel("Labor Force Participation Rate")
plt.ylabel("ED Visit Rate (Opioid-Related)")
plt.show()
```

```python
exploratory_data = basic_data.copy()

#Include unreliable rates
for idx,row in basic_data.iterrows():
    if row["Crude Rate"] == "Unreliable":
        exploratory_data.at[idx,"Crude Rate"] = (float(basic_data.at[idx,"Crude Rate Lower 95% Confidence Interval"]) + float(basic_data.at[idx,"Crude Rate Upper 95% Confidence Interval"]) / 2)

#Drop Supressed Values
exploratory_data = exploratory_data[exploratory_data["Crude Rate"] != "Suppressed"]

exploratory_data["Crude Rate"] = exploratory_data["Crude Rate"].astype(float)


#Compare ED Visit Rate and Death Rate

plt.scatter(exploratory_data["ED Visit Rate"],exploratory_data["Crude Rate"],c=exploratory_data["Year"])
plt.title("ED Visit Rate vs. Death Rate")
plt.xlabel("ED Visit Rate (Opioid-Related)")
plt.ylabel("Opioid Death Rate")
plt.show()

plt.scatter(exploratory_data["Crude Rate"],exploratory_data["ED Visit Rate"],c=exploratory_data["Year"])
plt.title("Death Rate vs. ED Visit Rate")
plt.ylabel("ED Visit Rate (Opioid-Related)")
plt.xlabel("Opioid Death Rate")
plt.show()
```

## Tensorflow ED Visit Rate Regression Model

```python
tf.random.set_seed(12)

#Create training, test sets
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(basic_data[basic_features], basic_data["ED Visit Rate"], test_size=0.2, random_state=12)
```

```python
sns.pairplot(train_data[basic_features],diag_kind="kde")
```

```python
train_stats = train_data.describe()
train_stats = train_stats.transpose()
train_stats
```

```python
#Feature Normalization
def norm(x):
    return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_data)
normed_test_data = norm(test_data)
```

```python
def build_model():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(64,activation='relu',input_shape=[len(train_data.keys())]),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(1)
    ])
        
    optimizer = tf.keras.optimizers.RMSprop(0.001)
        
    model.compile(loss='mse',optimizer=optimizer,metrics=['mae','mse'])
    
    return model
```

```python
model = build_model()
model.summary()
```

```python
display(normed_train_data)
display(train_labels)
```

```python
EPOCHS = 1000

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(normed_train_data, train_labels, epochs=EPOCHS, validation_split = 0.15, verbose=0, callbacks=[early_stop])
```

```python
hist = pd.DataFrame(history.history)
#display(hist)

plt.plot(hist["loss"])
plt.plot(hist['val_loss'],color='orange')
plt.legend()
plt.ylabel("Loss")
plt.xlabel("Epoch")

plt.show()
```

```python
#Make Predictions
test_predictions = model.predict(normed_test_data).flatten()

a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [Death Rate]')
plt.ylabel('Predictions [Death Rate]')
lims = [0, 50]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```

```python
error = test_predictions - test_labels
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [Death Rate]")
_ = plt.ylabel("Count")
```

We can see that this model is not as accurate as the death rate model, likely as a result of less data with greater variance.

```python
#Evaulate accuracy
print('R^2: ', sklearn.metrics.r2_score(test_labels,test_predictions))
print('MSE: ', sklearn.metrics.mean_squared_error(test_labels,test_predictions))
print('Explained Variance: ', sklearn.metrics.explained_variance_score(test_labels,test_predictions))
```
